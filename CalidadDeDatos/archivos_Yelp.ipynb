## DF Businees  .pkl

### parace tener duplicadas sus columnas


```python
# eliminar las columnas duplicadas  
df_pkl = df_pkl.loc[:, ~df_pkl.columns.duplicated()]
```

```python
df_pkl.head()
```

```python
df_pkl.info()
```

### % de data nula en cada columna 

```python
for column in df_pkl.columns:
    percentage = df_pkl[column].isnull().mean()  
    print(column + ': ' + str(round(percentage*100, 2))) 
```

* tanto 'attributes' como 'hours' tiene data nula  y ademas son columnas anidadas  

*  ademas a considerar que 'attributes'  tiene data variada

## DF reviews.json

* no se puede cargar como dataframe ya que satura la memeria ram  por su tamaÃ±o 

```python
file_path = "yelp/review.json"

with open(file_path, "r") as json_file:
    for line in json_file:
        review = json.loads(line)
        
        # Imprimir los valores en columnas
        print(f"Review ID: {review['review_id']}")
        print(f"User ID: {review['user_id']}")
        print(f"Business ID: {review['business_id']}")
        print(f"Stars: {review['stars']}")
        print(f"Useful: {review['useful']}")
        print(f"Funny: {review['funny']}")
        print(f"Cool: {review['cool']}")
        print(f"Text: {review['text']}")
        print(f"Date: {review['date']}")
        
        # Imprimir una lÃ­nea separadora para mayor claridad
        print("-" * 50)
```

* otra alternativa   

```python
with open(file_path, "r") as json_file:
    for line in json_file:
        review = json.loads(line)
        print(review)
```
