```python
import pandas as pd
import json
```

## DF Businees  .pkl

```python
df_pkl = pd.read_pickle('yelp/business.pkl')
```

```python
df_pkl.tail()
```

### calidad de datos 

```python
df_pkl.info()
```

### parace tener duplicadas sus columnas


```python
# eliminar las columnas duplicadas  
df_pkl = df_pkl.loc[:, ~df_pkl.columns.duplicated()]
```

```python
df_pkl.head()
```

```python
df_pkl.info()
```

### % de data nula en cada columna 

```python
for column in df_pkl.columns:
    percentage = df_pkl[column].isnull().mean()  
    print(column + ': ' + str(round(percentage*100, 2))) 
```

* tanto 'attributes' como 'hours' tiene data nula  y ademas son columnas anidadas  

*  ademas a considerar que 'attributes'  tiene data variada
