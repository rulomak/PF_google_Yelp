##  genero una funcion para hacerlo de manera mas eficiente

```python
   

def procesar_archivos_json(archivos):
    resultados = []
    for archivo in archivos:
        with open(archivo) as json_file:
            for line in json_file:
                data = json.loads(line)
                resultados.append(data)
    
    df = pd.DataFrame(resultados)
    return df
```

* creo la lista de archivos a procesar,  mediante un for  

```python
lista_archivos = []
for i in range(12):
    lista_archivos.append(f'alabama/{i+1}.json')
    
```

```python
lista_archivos
```

### genero el dataframe completo de alabama  

```python
df_alabama_complete= procesar_archivos_json(lista_archivos)
```
